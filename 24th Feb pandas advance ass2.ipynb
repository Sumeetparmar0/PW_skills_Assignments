Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
df1 = pd.DataFrame({'A': [10, 20, 30], 'B': [40, 50, 60], 'C': [70, 80, 90]})
```

```python
df1
```

```python
def reindex_dataframe(df):
    new_index = range(1, 2 * len(df) + 1, 2)
    df_reindexed = df.set_index(pd.Index(new_index))
    return df_reindexed
```

```python
df_reindexed = reindex_dataframe(df1)
```

Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.
For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
calculate and print the sum of the first three values, which is 60.

```python
df3 = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
```

```python
df3
```

```python
def calculate_sum(df):
    if 'Values' in df.columns:
        first_three_values = df['Values'].iloc[:3]
        sum_of_first_three_values = first_three_values.sum()
        print(f"Sum of the first three values in 'Values' column: {sum_of_first_three_values}")
```

```python
calculate_sum(df3)
```

Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,

```python
def Calc_Stat(df, Column):
    Mean = df[Column].mean()
    Median = df[Column].median()
    STD = df[Column].std()
    print(f"Mean of '{Column}':{Mean}")
    print(f"Median of '{Column}':{Median}")
    print(f"Standard deviation of '{Column}':{STD}")
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.

```python
df9 = pd.DataFrame({
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05', '2022-01-06', '2022-01-07'],
    'Sales': [10, 15, 20, 25, 30, 35, 40]
})
```

```python
df9
```

```python
def calculate_moving_average(df, value_column, date_column, window_size):
    if value_column in df.columns and date_column in df.columns:
        # Convert the 'Date' column to datetime format
        df[date_column] = pd.to_datetime(df[date_column])

        # Sort DataFrame by 'Date'
        df.sort_values(by=date_column, inplace=True)

        # Calculate the moving average
        df['MovingAverage'] = df[value_column].rolling(window=window_size, min_periods=1).mean()

        # Print the DataFrame with the new column
        print(df)
```

```python
calculate_moving_average(df9, 'Sales', 'Date', window_size=7)
```

```python
df9
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
df11 = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-15', '2023-01-25', '2023-02-05', '2023-02-10']
})
```

```python
df11
```

```python
def select_rows_between_dates(df, date_column, start_date, end_date):
    if date_column in df.columns:
        df[date_column] = pd.to_datetime(df[date_column])
        mask = (df[date_column] >= start_date) & (df[date_column] <= end_date)
        selected_rows = df[mask]
        return selected_rows
```

```python
selected_rows = select_rows_between_dates(df11, 'Date', '2023-01-01', '2023-01-31')
print(selected_rows)
```
