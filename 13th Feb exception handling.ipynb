Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

ANS:The LookupError class is a base class for exceptions that are raised when a sequence or mapping (like a list, tuple, dictionary, or string) is accessed with an invalid key or index. Two common exceptions that are derived from LookupError are KeyError and IndexError.

KeyError:A KeyError is raised when you attempt to access a dictionary using a key that doesn't exist in the dictionary.

```python
my_dict = {"apple": 5, "banana": 3, "cherry": 8}

try:
    count = my_dict["grape"]
except KeyError as e:
    print("Error:", e)
```

IndexError:
An IndexError is raised when you attempt to access a sequence (e.g., a list or a tuple) with an index that is out of range

```python
my_list = [10, 20, 30]

try:
    value = my_list[3]  
except IndexError as e:
    print("Error:", e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ANS: Both exceptions in Python that occur when there are issues related to importing modules. 
ImportError is a base class for exceptions raised when an import statement fails for some reason. This exception can be raised in various situations.
ModuleNotFoundError is a specific exception. It is derived from ImportError and is raised when an import statement cannot find the module to be imported. it provides a more precise error message when a module is not found.
