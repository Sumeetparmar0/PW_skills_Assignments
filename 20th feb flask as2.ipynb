Q1. Explain GET and POST methods.

ANS:
GET Method:

The GET method is one of the HTTP methods used to request data from a specified resource. When a client sends a GET request to a server, it is asking the server to retrieve or return data. This method is often used for simple queries where the parameters are appended to the URL. For example, when you submit a search query through a web browser, it is typically sent as a GET request.

POST Method:

The POST method is another HTTP method, but it is used to submit data to be processed to a specified resource. Unlike GET, the data is not appended to the URL but is included in the body of the request. POST requests are commonly used when submitting forms on the web or when uploading a file.

n summary, the GET method is used for requesting data, and the POST method is used for submitting data to be processed.

Q2. Why is request used in Flask?

ANS:
In Flask, the request object is used to access incoming request data. It allows a Flask application to handle data sent by a client (usually a web browser) as part of an HTTP request. The request object provides convenient methods and attributes to access various parts of the request, such as form data, query parameters, cookies, and more.
The request object is a crucial component in building dynamic web applications with Flask, allowing the server to interact with the data sent by clients during HTTP requests.

Q3. Why is redirect() used in Flask?

ANS:
In Flask, the redirect() function is used to redirect the client to a different URL. It is a part of the flask module and is particularly useful when you want to send the user to another route or an external URL after processing a request. This is commonly used in scenarios like form submissions, login/logout actions, or any situation where you want the user to be directed to a different page.

Q4. What are templates in Flask? Why is the render_template() function used?

ANS:
In Flask, templates are used to separate the presentation logic from the application logic. They allow you to build HTML pages with placeholders for dynamic content that can be filled in by your Python code. Templates are often used to generate HTML dynamically, incorporating data from the server-side into the final HTML sent to the client.

The render_template() function in Flask is used to render HTML templates. It takes the name of the template file as an argument, along with any additional data that should be passed to the template. This function automatically looks for template files in a folder named templates in the root directory of your Flask project.
