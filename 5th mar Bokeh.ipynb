Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, show

# Create data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 7]

# Create a figure
p = figure(title="Scatter Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a scatter plot
p.circle(x, y, size=10, color="navy", alpha=0.5)

# Show the plot
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
ANS:
In Bokeh, glyphs are visual shapes that you can add to a plot to represent your data. Glyphs can be lines, circles, squares, patches, and more. You can add glyphs to a Bokeh plot using the figure function.

```python
#here is the example
from bokeh.plotting import figure, output_notebook, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title="Line Glyph Example", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line glyph to the figure
p.line(x, y, legend_label='Line', line_width=2, line_color="blue")

# Show the plot
output_notebook()
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

ANS:
A Bokeh server is a Python process that runs Bokeh applications, which are interactive data visualization web applications. Bokeh applications can be defined and run on a Bokeh server to create interactive plots that can be updated in real time. The Bokeh server allows for both simple and complex applications where the plot's data can change dynamically based on user interactions or external events.

Define a Bokeh Application:
Create a function or a class that defines your Bokeh plot and its interactivity.
This function or class should return a Bokeh layout (e.g., column, row) containing the plots and widgets.

Run the Bokeh Server:
Use the bokeh serve command to run the Bokeh server, specifying the script or module containing your Bokeh application.

Update and Modify Data Dynamically:
With the Bokeh server running, you can dynamically update the data in response to user interactions or external events.
Bokeh provides mechanisms like ColumnDataSource for efficient and reactive data handling.
