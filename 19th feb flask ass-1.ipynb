Q1. What is Flask Framework? What are the advantages of Flask Framework?

ANS:
Flask is a micro web framework for Python designed to be simple and lightweight. It is classified as a microframework because it does not require particular tools or libraries. Flask supports extensions to add additional functionality, making it highly modular and adaptable. It is widely used for developing web applications and APIs due to its simplicity and ease of use.

Advantages of Flask Framework:
Simplicity: Flask follows a minimalist philosophy, providing the essentials to get started without imposing unnecessary complexity. This simplicity makes it easy for developers to understand and use.

Flexibility: Flask is unopinionated, allowing developers to choose the tools and libraries they prefer. This flexibility makes it adaptable to a variety of projects and coding styles.

Extensibility: Flask is designed to be extended through a variety of plugins and extensions. Developers can add specific functionalities as needed, promoting a modular and scalable development approach.

Built-in Development Server: Flask comes with a built-in development server, making it easy to test and debug applications during development without requiring additional server configurations.

RESTful Request Dispatching: Flask supports RESTful request dispatching, making it suitable for building APIs and services that follow REST principles.

Scalability: While Flask is often used for small to medium-sized projects, it can be used for larger applications as well. The modular design allows developers to scale up their applications as needed.

Q3. What is App routing in Flask? Why do we use app routes?

ANS:

App Routing in Flask:

App routing in Flask refers to the mechanism by which URLs are mapped to specific functions or views within a Flask application. It defines how different endpoints or routes of a web application correspond to specific functionalities or responses. In Flask, app routing is achieved using the @app.route() decorator, which binds a function to a URL.

Why do we use app routes?

developers can create a structured and modular design where each route corresponds to a particular functionality or page. This approach enhances code organization, making it easier to maintain and troubleshoot. App routes also support dynamic URLs, facilitate the handling of various HTTP methods, and contribute to the creation of RESTful APIs. Additionally, the routing system enables URL building and fosters a clear separation of concerns within a Flask application, ultimately providing a flexible and intuitive way to structure web projects.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

ANS:
In Flask, the url_for() function is used for URL building. It generates a URL for a specified endpoint (a function decorated with @app.route()).

syntex is as given below for flask


from flask import Flask, url_for

app = Flask(__name__)

Route for "/welcome"
@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

Route for "/"
@app.route('/')
def company_info():
    # Using url_for() to generate a URL for the 'welcome' endpoint
    welcome_url = url_for('welcome')
    
    return f'''
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999
    
    Welcome Message URL: {welcome_url}
    '''

if __name__=="__main__":
    app.run(host="0.0.0.0")
