Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.


import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.


```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()

#Scatter Plot: Displays individual data points as markers on a 2D plane, useful for examining relationships between two variables.
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

ANS: The subplot() function in Matplotlib is used to create a grid of subplots in a single figure. It takes three arguments: the number of rows, the number of columns, and the index of the subplot. This allows you to create and arrange multiple plots in a single figure.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.array([0, 1, 2, 3, 4, 5])

# Subplot 1
plt.subplot(2, 2, 1)
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1)
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2)
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3)
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4)
plt.title('Line Plot 4')


# Show the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

ANS: 
A bar plot is a graphical representation of data where individual bars represent different categories or groups. It is commonly used to compare and display the values of different categories. Bar plots can be either vertical or horizontal.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
#instead of bar if we use barh then plot will be horizontal bar
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Bar Plot: Company vs. Profit')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

ANS:
A box plot, also known as a whisker plot, displays the summary of a set of data values. It shows the distribution and spread of data, including minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It is useful for identifying outliers and understanding the overall distribution of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data_to_plot = [box1, box2]
plt.figure(figsize=(8, 5))
plt.boxplot(data_to_plot, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot: Box 1 vs. Box 2')
```
